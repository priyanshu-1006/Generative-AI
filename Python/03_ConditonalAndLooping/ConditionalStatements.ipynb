# Conditional Statement

#### Conditional statements works like real life scenerio where if a certain condition is meeting then a statement of code or a block of a code gets executed

```python
# Eligibility for driving on meeting a certain conditon
## Conditional Statements--> If statements where if a condition is true then only the code will be executed.

Age=int(input("Enter your age: "))
if Age>=18:
    print("You are eligible for driving")
else:
    print("You are not eligible for driving")   
    
```

## Conditional Statements
1. if statement
2. elif statement
3. else statement

```python
## Figure out that who is the tallest among the three friends

Amit=int(input("Enter the height of Amit in cm: "))
Rahul=int(input("Enter the height of Rahul in cm: "))
Suresh=int(input("Enter the height of Suresh in cm: "))

if Amit>Rahul and Amit>Suresh:
    print("Amit is the tallest among the three friends")    
elif Rahul>Amit and Rahul>Suresh:
    print("Rahul is the tallest among the three friends")               
else:
    print("Suresh is the tallest among the three friends")
```

## Nested Statement

Nested Conditional Statements involve if statements inside another if statement

```python
## Compare the height of three friends and figure out who is the tallest among them. If there are two or more friends with the same height, then print that they are of the same height.

Amit_height=int(input("Enter the height of Amit in cm: "))
Rahul_height=int(input("Enter the height of Rahul in cm: "))
Suresh_height=int(input("Enter the height of Suresh in cm: "))

if Amit_height>Rahul_height and Amit_height>Suresh_height:
    print("Amit is the tallest among the three friends")            
elif Rahul_height>Amit_height and Rahul_height>Suresh_height:
    print("Rahul is the tallest among the three friends")
elif Suresh_height>Amit_height and Suresh_height>Rahul_height:
    print("Suresh is the tallest among the three friends")
elif Amit_height==Rahul_height and Amit_height>Suresh_height:
    print("Amit and Rahul are of the same height and taller than Suresh")
elif Amit_height==Suresh_height and Amit_height>Rahul_height:
    print("Amit and Suresh are of the same height and taller than Rahul")
elif Rahul_height==Suresh_height and Rahul_height>Amit_height:
    print("Rahul and Suresh are of the same height and taller than Amit")
else:
    print("All three friends are of the same height")               
```

```python
# Whether if the current year is a leap year or not

year=int(input("Enter the year: "))
if (year%4==0 and year%100!=0) or (year%400==0):
    print(f"{year} is a leap year")     
else:    print(f"{year} is not a leap year")    
```
