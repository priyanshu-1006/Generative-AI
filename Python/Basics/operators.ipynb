## Operators

Operators helps in applying operations on the given operands

## Types of Operators   
1. Arithmetic Operators

Addition (+)
Subtraction (-) 
Product or Multiplication (*)
Division (/)
Modulus (%)
Floor (//)
Exponent (a**b)

2. Comparison Operators

Greater than (>)
Less than (<)
Greater than or equal to (>=)
Less than or equal to (<=)
Equality check (==)
not equal to (!=)

3. Logical Operators

and---> whether if both thw conditons are true or not
or--->whether one of two conditions true
not--->  inverse of result
