## Range()


range() function can be used for working as our iterable

range() takes an input parameter and the first parameter that the range will be taking as the input will be the initial value that you want to print or you want to operate on and then you will be passing the next inout after comma as the parameter as the final value and it can take third parameter as well and that will be the step size and that is the difference between the iterating values

```python
# print the numbers from 1 to 50

for i in range(1,51,1):
    print(i)
```

```python
# print the numbers from 1 to 50

for i in range(1,51,2):
    print(i)
```

```python
# Numbers from 0 to 100

for i in range(0,101,5):
    print(i)
```

In the range there is never a neccessity for mentioning the step size and similarily if you want to print the numbers or you to operate on numbers again there is no neccessity for passing the initial value as well

#### By using range() we can represent the numbers in a decreamenting manner or in a decreasing order as well

```python
# Print the numbers from 50 to 1 with a gap of 1 value in a decrementing order

for i in range(50,0,-1):
    print(i)
```

```python
# Print the numbers from 50 to 1 with a gap of 5 value in a decrementing order

for i in range(50,0,-5):
    print(i)
```

# While Loop

While loop is being used for iterating over a series of numbers when a condition is True and we have to initialise a value as well while iterating

```python
## By making the use of while loop we need to print numbers from 0 to 50

i=0
while i<=50:
    print(i)
    i+=1
    
```

```python
# Print even numbers from 0 to 50 using while loop

for i in range(51):
    if i%2==0:
        print(i)
```

# Loop Control Statements

#### Loop Control Statements are being used for terminating a loop in between when a condition is met or if you want to have some control over your iterating process or while iterating over the series of numbers

1. break ---> Break is our first loop control statement which will help you out to terminate the loop when a certain condition is met

```python
## Print the numbers from 0 to 50 but once number = 18 then terminate the loop or do not print
## either 18 or any number forward

for i in range(51):
    if i==18:
        break
    print(i)   
```

2. continue --->  continue is our second loop control statement which helps us in iterating over a series of numbers but helps us with skipping the values in between once we meet a condition

```python
## Print even numbers from 0 to 50

for i in range(51):
    if i%2==0:
        print(i)
    else:
        continue
```

```python
# Print the numbers from 1 to 20 and skip 17

for i in range(21):
    if i==17:
        continue
    print(i)
else:
    print("Loop is ended")
```

3. Pass ---> Pass statement is being used for doing nothing

```python
# Print the numbers from 0 to 20 but once 17 is occuring do nothing

for i in range(21):
    if i==17:
        print("Our program will do nothing when 17 is occuring")
        pass
    else:
        print(i)
```

### Nested loops
It is a loop inside another loop. The inner loop will be executed one time for each iteration of the outer loop.


```python
print(0,0)
print(0,1)
print(0,2) 
print(1,0)
print(1,1)
print(1,2)
```

### f string ---> f string will be helping us out with printing the numbers with the variable value present inside the colons so we can directly represent the variable value inside the colons message itself where variable value will be shown inside the {}

```python
# Nested loop can help us out in iterating over a series of number as with the proper combination of iterating numbers

for i in range(6):
    for j in range(4):
        print(f"i value is: {i} and j value is: {j}")
```

```python
## Print the sum of first N Natural numbers
n = int(input("Enter the value of N: "))
sum = 0
for i in range(1, n + 1):
    sum = sum + i
print(f"The sum of first {n} natural numbers is: {sum}" )
```

```python
## Check Whether a number is prime or not

number = int(input("Enter a number: "))
if number > 1:
    for i in range(2, number):
        if number%i==0:
            print(f"{number} is not a prime number")
            break
    else:
        print(f"{number} is a prime number")

            
```

```python
# Print Prime numbers from 0 to 100

for num in range(101):
    if num > 1:
        for i in range(2, num):
            if num%i==0:
                break
        else:
            print(num)
```
