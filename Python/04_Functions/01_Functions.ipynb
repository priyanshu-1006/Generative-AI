## Recursive Function

By a Recursive Function we can say that a function is going to call itself again and again

```python
## Factorial of a number (by iteration)

def factorial(num):
    if num==0 or num==1:
        return 1
    else: 
        factorial = 1
        for i in range(1,num+1):
            factorial = factorial*i
        print(f"The factorial of {num} is: {factorial}")
```

```python
number=int(input("Enter a number to find its factorial: "))
factorial(number)
```

```python
def factorial_recursion(num):
    if num==0 or num==1:
        return 1
    else:
        return num*factorial_recursion(num-1)
```

```python
factorial_recursion(5)
```

```python
factorial_recursion(1)
```

```python
## return num* factorial_recursion(num-1)
'''
base condition ---> which will stop the loop or loop or which will let the function to stop calling itself 
In our case, base condition was ---> if i == 0 or i==1 ---> 1
num* factorial_recursion(num-1) 
num==5
5* factorial_recursion(4)--->
5* 4* factorial_recursion(3)--->
5* 4* 3* factorial_recursion(2)--->
5* 4* 3* 2* factorial_recursion(1)

'''
```

```python
input_recursive=int(input("Enter a number to find its factorial: "))
print(f"The factorial of {input_recursive} is: {factorial_recursion(input_recursive)}")
```

## Function with Variable Length Arguments

### Positional Arguments
Position arguments are being used so that we can pass multiple values inside our function as an argument irrespective of the number of values we have mentioned as the parameter inside our function

```python
## function to add multiple numbers irrespective of the number of numbers that we want to add

def addition_positional(*num):
    add=0
    for i in num:
        add=add+i
    print(f"The addition of {num} is: {add}")    
```

```python
addition_positional(5, 10, 15, 20)
```

```python
addition_positional(23,44,53,545,545,45,45,45,45,45,45,45,45,45,45,45,45,45)
```

### Keyword Arguments
Takes the input as the keyword values and irrespective of any number of keyword values you are going to pass it wont matter

```python
def data(**user_data):
    for key, value in user_data.items():
        print(f"{key}: {value}")
```

```python
data(name="Priyanshu", age=22, city="Varanasi", profession="Data Scientist")
```

You guys want to pass keyword value pairs as your input along with positional arguments as well

```python
def variable(*positional, **keyword):
    for i in positional:
        print(f'Positonal Values are : ',i)
    for key,value in keyword.items():
        print(f'Keyword Values are : {key} and value is : {value}')    
```

```python
variable(1,2,3,4,5, name="Priyanshu", age=22, city="Varanasi", profession="Data Scientist")
```
