# Map()

#### Map function applies a given function operation to All items in an iterable or a List and returns a Map Object (an Iterator)

-------------------------------------

#### Map function is particularly useful for Transforming data or applying an operation on all items of the iterable

----------------------------------------

#### Map() is going to help us out with avoiding unnecessary lines of code and we can apply the operation on all values or all items in an iterable and this is the sign of a good programmer

------------------------------------

#### Syntax for Map()

list(map(function,iterable))

```python
## Squaring all items of a list 
## list -----> [1,4,5,6,23,45,67,89,90]
## []---> [1,16,25,36,529,2025,4489,7921,8100]

def square(list_items):
    empty_list = []
    for item in list_items:
        squared_item = item**2
        empty_list.append(squared_item)
    print(empty_list)
```

```python
square([1,4,5,6,23,45,67,89,90])
```
