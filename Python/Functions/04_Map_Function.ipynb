# Map()

#### Map function applies a given function operation to All items in an iterable or a List and returns a Map Object (an Iterator)

-------------------------------------

#### Map function is particularly useful for Transforming data or applying an operation on all items of the iterable

----------------------------------------

#### Map() is going to help us out with avoiding unnecessary lines of code and we can apply the operation on all values or all items in an iterable and this is the sign of a good programmer

------------------------------------

#### Syntax for Map()

list(map(function,iterable))

```python
## Squaring all items of a list 
## list -----> [1,4,5,6,23,45,67,89,90]
## []---> [1,16,25,36,529,2025,4489,7921,8100]

def square(list_items):
    empty_list = []
    for item in list_items:
        squared_item = item**2
        empty_list.append(squared_item)
    print(empty_list)
```

```python
square([1,4,5,6,23,45,67,89,90])
```

```python
## Similar problem statement with a Map() --->

print(list(map(lambda x:x**2,[1,4,5,6,23,45,67,89,90])))
```

```python
### Map function with basic function in it

def square_numbers(num):
    return num**2

print(list(map(square_numbers,[1,4,5,6,23,45,67,89,90])))    
```

### Whenever you are going to pass the general function inside your map function as the parameter than you dont have to use parenthesis

#### While Mapping the value of one iterable with another iterable we can actually apply the operation in between the valus of the two iterables in a synchronous way 
#### [1,2,3,4] <-----> [5,6,8,2]---> Addition---> [6,8,11,6]

```python
## Map Function with Multiple Iterables

numbers_1=[1,4,5,8]
numbers_2=[8,5,6,7]

print(list(map(lambda x,y:x+y,numbers_1,numbers_2)))
```

```python
### Map() to convert all string values items in an iterable into an integer

print(list(map(int, ['1', '2', '9', '6', '5','8'])))
```

```python
map_values=list(map(int, ['1', '2', '9', '6', '5','8']))
print(map_values)
```

```python
employee_name=['Priya','Rohit','Sonia','Rahul','Riya','Rohini','Satyarth','Satyam']
print(list(map(str.upper,employee_name)))
```

```python
### Retrieve the values inside a dictionary using Map()

employees=[{'Name':'Priya','Age':24,'Qualification':'B.Tech'},
           {'Name':'Rohit','Age':25,'Qualification':'B.Tech'},
           {'Name':'Sonia','Age':24,'Qualification':'B.Tech'},
           {'Name':'Rahul','Age':26,'Qualification':'B.Tech'},
           {'Name':'Riya','Age':24,'Qualification':'B.Tech'},
           {'Name':'Rohini','Age':25,'Qualification':'B.Tech'},
           {'Name':'Satyarth','Age':27,'Qualification':'B.Tech'},
           {'Name':'Satyam','Age':28,'Qualification':'B.Tech'}]
```

```python
def age_retriever(employee):
    return employee['Age']
```

```python
print(list(map(age_retriever,employees)))
```
