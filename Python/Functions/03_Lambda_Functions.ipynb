# Lambda Functions

#### Lambda Functions are the the One Linear functions defined by the Lambda Keyword

#### Syntax for Lambda Function:

Lambda Arguments:expression

 #### 1. lambda --> Keyword which simply let the python interpreter know that the current function is our lambda function

 #### 2. Arguments --> Input values for our Function

 #### 3 Expression --> Logic (Function Body)

 

lambda x:x**2

-------------------------------------------------------

### They can have any number of Arguments but only one expression or only one logic 
### They are used for Short Operations or as Arguments to Higher Order Functions

### It can be treated as a Function without a Name and they Dont even require a return statement

```python
## function which calculates the square of the numbers

def square(num):
    return num**2
```

```python
print(square(5))
```

-------------------------------------------------------

```python
## This similar problem can be solved using lambda function as well.

lambda_function = lambda num: num**2
```

```python
type(lambda_function)
```

```python
lambda_function(5)
```

----------------------------------------

```python
## Lambda Function to calculate the sum of two numbers

sum=lambda x,y:x+y
print(sum(5,10))
```

--------------------------------------

```python
product = lambda x,s,t,y:x*y*t*y
print(product(5,10,32,3))
```
