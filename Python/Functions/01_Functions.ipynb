```python
## We are creating a Function to see that how the translation is going to be their
# from English to Hindi language

def translator(English_text):
    """This function will take English text as input and will return the Hindi translation of that text."""

    if English_text=="Hello":
        print(f"The translation of {English_text} in Hindi is: नमस्ते")

    if English_text=="Help":
        print(f"The translation of {English_text} in Hindi is: मदद")

    if English_text=="required" or English_text=="requirement":
        print(f"The translation of {English_text} in Hindi is: आवश्यक")

    
```

## For having the access to our Function and to provide the input to our function we can simply call the function

## Function calling is just like calling a waiter in a restaurant where in just like we call the waitor and provide the input as the food item and then waitor provide us with the food itself, like in a similar way we can call a function as well and can provide the input and can get the required output after calling the function only

```python
## calling the function --> Mention the name of the function and then provide the input in the parenthesis
English_input=input("Enter the English text that you want to translate to Hindi: ")
translator(English_input)
```
