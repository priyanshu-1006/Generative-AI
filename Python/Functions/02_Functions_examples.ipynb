### Q1. To create a function that converts temperature from celcius to Fahrenheit or Fahrenheit to celsius

-------------------------------------------------------------------------

Answer

1 degree Celcius = 9/5 + 32
For ex-
    45 def C = 45*9/5 + 32 F

1 Fahrenheit = 5/9 - 32 Celcius
For ex-
    45 F = (45-32)*5/9 Celcius

```python
## solution

def convert_to_celsius(fahrenheit):
    celsius = (fahrenheit - 32) * 5.0/9.0
    print(f"{celsius} degrees Celsius")
fahrenheit = 100
celsius = convert_to_celsius(fahrenheit)

def convert_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9.0/5.0) + 32
    print(f"{fahrenheit} degrees Fahrenheit")
celsius = 37
fahrenheit = convert_to_fahrenheit(celsius)
```

```python
convert_to_celsius(212)
```

```python
convert_to_fahrenheit(100)
```

### Q2.  Define a function that would behave as a Password Strength Checker

---------------------------------------------------

```python
'''
Strong Password -->
if your Paswword:
    1. Should be atleast 8 characters long
    2. Should contain atleast one uppercase letter
    3. Should contain atleast one lowercase letter
    4. Should contain atleast one digit
    5. Should contain atleast one special character

Weak Password -->
if your Password does not match any of the above criteria, then it is a weak password.
'''
```

Solution

```python
# isupper() --> Checks that whether the string contains atleast one uppercase letter
# islower() --> Checks that whether the string contains atleast one lowercase letter
# isdigit() --> Checks that whether the string contains atleast one digit
```

```python
def password_strength(password):
    if len(password) < 8:
        print(f"The entered password '{password}' is a Weak Password")
    elif not any(char.isupper() for char in password):
        print(f"The entered password '{password}' is a Weak Password") 
    elif not any(char.islower() for char in password):
        print(f"The entered password '{password}' is a Weak Password")
    elif not any(char.isdigit() for char in password):
        print(f"The entered password '{password}' is a Weak Password")
    elif not any(char in "!@#$%^&*()-_=+[]{}|;:'\",.<>?/" for char in password):
        print(f"The entered password '{password}' is a Weak Password")
    else:
        print(f"The entered password '{password}' is a Strong Password")
```

```python
password_strength("Priyanshu@123")
```

```python
password_strength("Priyanshu@")
password_strength("priyanshu")
```

### Q3. Check that whether if the current string is a palindrome or not

-----------------------------------------------------

#### Palindrome String

Palindrome string is going to have characters similar to it from both the sides or even if you will reverse the string still you will be getting the given string itself

##### palindrom ---> aba, aaabbaaa.


--------------------------------------------------------

Solution

```python
## Palindrome Checker

def palindrome(string):

    string_value=string.lower().replace(" ","")
    
    if string_value == string_value[::-1]:
        print(f"The entered string '{string}' is a Palindrome")
    else:
        print(f"The entered string '{string}' is not a Palindrome")
```

```python
palindrome("ABA")
```

```python
palindrome("ABB")
```

```python
palindrome("AaBB aa")
```

```python
palindrome('A man a plan a canal Panama')
```

### Q4.  Calculate the Total Cost of items in a shopping cart

----------------------------------------------------

Solution

```python
cart = [{'item1':'Apple', 'price': 5, 'quantity': 2},
        {'item2':'Banana', 'price': 4, 'quantity': 5},
        {'item3':'Orange', 'price': 8, 'quantity': 3}]


## index in our cart list --> keyname
## total --> price * quantity

def cart_total(cart):
    total = 0
    for item in cart:
        total += item['price'] * item['quantity']
    print(total)
    
```

```python
cart_total(cart)
```
