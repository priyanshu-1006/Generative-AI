```python
## open() ---> Opening a file
## first Parameter ---> file name
## second parameter ---> mode of opening the file
## Mode ---> it decides the purpose of opening the file

## Mode ---> r ---> read mode
## Mode ---> w ---> write mode      
## Mode ---> r+---> read and write mode
## Mode ---> w+---> write and read mode
## Mode ---> a ---> append mode
```

## Reading your file
## read() ---> Reads the content from the file

```python
f=open('example.txt','r')
content=f.read()
print(content)
f.close()
```

-----------------------------------------------------

### Writing the Content into the File

##### write()--->  It overwrites the content into the text file
##### append()---> It adds the data from the end of the file

## write() ---> Helps us out in writing the content into the file

```python
f=open('example.txt','w')
f.write('Hello We are now Writing in the file')
f.close()
```

---------------------------------------------------

## Appending the content into our file

append()----> It adds data from end of the file


```python
## Append mode

f=open('example.txt','a')
f.write('This is the new line which is being added to the file')
f.close()
```

---------------------------------------------------

#### with ---> It will be used for first of all opening the file as a object and there is no requirement for closing file as well

```python
with open('example.txt','r') as f:
    content_data=f.read()
    print(content_data)
```

```python
with open('example.txt','w') as file:
    file.write('This is the new line which is being added to the file')
```

```python
with open('example.txt','a') as files:
    files.write('\nThis is the new line which is being added to the file')
```

```python
## Write the content into our file line by line --> writelines() method
```

```python
lines=['First Line\n','Second Line\n','Third Line\n']
with open('example.txt','w') as f:
    f.writelines(lines)
```

## Mode ---> Write the content into the binary file ---> wb
## Mode ---> Read the content from the binary file ---> rb

```python
data=b'\x00\x01\x02\x04'
with open('example.txt','wb') as binary_data:
    binary_data.write(data)
```

```python
with open('example.txt','rb') as binary_read:
    binary_content=binary_read.read()
    print(binary_content)
```

```python
## Reading the total lines, Total Words, Total Characters from a file

def count_lines(file):
    with open(file,'r') as files:
        lines=files.readlines()
        total_lines=len(lines)
        total_words=(len(line.split()) for line in lines)
        total_characters=sum(len(line) for line in lines)
    return total_lines,total_words,total_characters
file_name='Total_data.txt'
Total_lines,Total_words,Total_characters=count_lines(file_name) 
print(f'Total Lines: {Total_lines}')
print(f'Total Words: {Total_words}')   
print(f'Total Characters: {Total_characters}')
```

readlines()---> Reads the line in a way where it is like the lines are stored inside a list ['First_Line', 'Second_Line', 'Third_Line']

```python
## Write the content into a file and then after writing the content into that file we are
## Going to read the content from that file as well ---> seek() method and w+
```

```python
### w+ ---> First of all as W is there it will write the content into the file and then after that we are going to read the content from that file as well

### seek() method ---> it is used to move the file pointer to a specific position in the file. By default, when you open a file, the file pointer is at the beginning of the file. After writing to the file, the file pointer will be at the end of the file. If you want to read from the beginning of the file after writing, you need to use the seek() method to move the file pointer back to the beginning.
```

```python
with open('Total_data.txt','w+') as file:
    file.write('This is the new line which is being added to the file')
    file.seek(0)  ## Moving the file pointer back to the beginning of the file
    content=file.read()
    print(content)
```
