# OS Library

#### OS Library is able to interact with our Operating System and can let us work with Multiple Directories and files.

```python
## getcwd()--> It is capable of providing us with the idea about the current working directory 
## getcwd--> get current working directory
```

```python
import os 
```

```python
current_path = os.getcwd()
print("The current working directory is: ", current_path)
```

```python
## mkdir()--> It is used to create a new directory in the current working directory. It takes the name of the directory as an argument and creates it in the current working directory.
```

```python
new_path="New_Folder"
os.mkdir(new_path)
print("The new directory is created in the current working directory: ", new_path)
```

```python
## listdir()--> It is used to list all the files and directories in the current working directory. It takes the path of the directory as an argument and returns a list of all the files and directories in that directory.
```

```python
All_Items=os.listdir('.')
print("The list of all the files and directories in the current working directory is: ", All_Items)
```

```python
## Joining 2 paths together

### os.path.join()--> It is used to join two paths together. It takes two paths as arguments and returns a single path by joining them together.

dir_name="Folder_Data"
file_name="File_Data.txt"
Full_Path=os.path.join(dir_name,file_name)
print("The full path of the file is: ", Full_Path)
```

```python
## I want to join Multiple Paths with the current path

current_complete_path=os.path.join(os.getcwd(),dir_name,file_name)
print("The full path of the file is: ", current_complete_path)
```

------------------------------------

```python
## os.path.exists()--> It is used to check whether a file or directory exists in the specified path. It takes the path of the file or directory as an argument and returns True if it exists, otherwise it returns False.
```

```python
given_path='C:/Users/priya/OneDrive/Desktop/CODING/Generative AI Course/Python/FileHandling/New_Folder'
if os.path.exists(given_path):
    print("The given path exists.")
else:    
    print("The given path does not exist.")
```

--------------------------------------

```python
## os.path.isfile()--> It is used to check whether a given path is a file or not. It takes the path of the file as an argument and returns True if it is a file, otherwise it returns False.

## os.path.isdir()--> It is used to check whether a given path is a directory or not. It takes the path of the directory as an argument and returns True if it is a directory, otherwise it returns False.
```

```python
path='New_Folder'

if os.path.isfile(path):
    print("The given path is a file.")
elif os.path.isdir(path):
    print("The given path is a directory.")
else:
    print("The given path is neither a file nor a directory.")    
```

```python
## The Complete Path ----> Absolute Path ---> os.path.abspath()

## os.path.abspath()--> It is used to get the absolute path of a file or directory. It takes the path of the file or directory as an argument and returns the absolute path of that file or directory.
```

```python
file_name="Total_Data.txt"  ## total_data.txt <---> relative path
complete_path=os.path.abspath(file_name) ## absolute path
print("The complete path or absolute path of the file is: ", complete_path)
```
