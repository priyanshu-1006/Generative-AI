```python
tuple_values=(1,2,3,4,5)
print(tuple_values)
```

```python
Nested_List[0][1]
```

```python
# Tuples inside the Nested List

tuple_list=[(1,2),(3,4,5),(6,7,8),(11,34,56,91,9)]
```

```python
tuple_list[0][:]
```

```python
# Nested Tuples

Nested_tuple=((1,2),(3,4,5),(6,7,8),(11,34,56,91,9))
```

```python
Nested_tuple[0][:]
```

```python
# Iteraating over a tuple inside a tuple

for i in Nested_tuple:
    for j in i:
        print(j, sep=" ")
    print(" ")
```
