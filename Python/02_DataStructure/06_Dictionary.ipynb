## Nested Dictionary
#### Nested Dictionary will be used for keeping a dictionary inside another dictionary

```python
## I want to store the information for multiple students like there name, marks, grade

Nested_dictionary={"Student_1":{"Name":"Reena","Marks":85,"Grade":"A"},
                   "Student_2":{"Name":"Riya","Marks":90,"Grade":"A+"},
                   "Student_3":{"Name":"Priyanshu","Marks":80,"Grade":"B+"}}
```

#### Iterate over these Nested Dictionary

```python
for outer_key,outer_value in Nested_dictionary.items():
    print(f"outer Key is : {outer_key} and outer Value is : {outer_value}")
    for key_1,value_1 in outer_value.items():
        print(f"inner Key is : {key_1} and inner Value is : {value_1}") # to get all the keys of the nested dictionary
```

## Dictionary Comprehension

Dictionary comprehension help us with using looping statement along with conditional statement inside our dictionary.

Syntax for Dictionary Comprehension
{expression for item in iterable if condition}

```python
comprehensive={x**2 for x in range(11)}
```

```python
 comprehensive
```

```python
comprehensive={x:x**2 for x in range(11)} # to create a dictionary using dictionary comprehension}

```

```python
comprehensive
```

```python
even_dictionary={x:x**2 for x in range(11) if x%2==0} # to create a dictionary using dictionary comprehension with condition
even_dictionary
```
