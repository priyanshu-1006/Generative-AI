## List Methods

1. Append()--> Insert the value at the end of th list
2. insert()--> Insert the value at a specific index location inside our list
3. remove()--> Remove the value from our list that we are going to pass inside it
4. pop()--> remove the value from the list at a specific index location
5. index()--> it provides the index of the value that we are going to pass inside it if that value is present inside the list 
6. sort()--> sort is going to help us out with sorting the values in as ascending order 
7. reverse()--> it can reverse the values present inside our list i.e., value at last index will be at first index now
8. count()--> It counts the number of occurence of a specific value that we are going to pass inside our list andit returns that count of occurence
9. clear()--> for removing all of the values that we are storing inside our list

## List Comprehension   

We are going to use Looping statements inside the list so that we can store the values the inside the list in an iterable format and in list comprehension conditional statements are allowed as well

## Syntax for List Comprehension is

### [expression for item in iterable if condition]

1. expression --> what we want to print
2. for item in iterable --> looping statement
3. if condition --> conditional statement

```python
# I need to store values from 1 to 50 inside the list
```

```python
list_items=[x for x in range(1,51)]
```

```python
list_items
```

```python
list_squared=[items**2 for items in range(1,16)]
```

```python
list_squared
```

```python
# if I am only interested in looking over the square value which are even

even_items=[x**2 for x in range(1,19) if x%2==0]
```

```python
even_items
```
