# Operator Overloading

```python
### Operator Overloading allows us to define how operators (+,-,/,*. etc) work with user objects. 

## Why is Operator Overloading useful ?

## Makes custon objects behave like built in types (numbers, strings, lists)

## Improves code readabililty by using natural syntax (object1 + object2) instead if object1.add(object2)
```

----

### Operator ---->
### + ---> Method name --> __add__(self,other)
### - ---> Method name --> __sub__(self,other)
### * ---> Method name --> __mul__(self,other)
### / ---> Method name --> __truediv__(self,other)
### == ---> Method name --> __eq__(self,other)
### > ---> Method name --> __gt__(self,other)


---

```python
class Avenger:

    def __init__(self, name, power):

        self.name = name

        self.power = power

    def __add__(self, other):

        return Avenger(self.name + " & " + other.name, self.power + other.power)

    def __sub__(self, other):

        return Avenger(self.name + " - " + other.name, self.power - other.power)

    def __mul__(self, other):

        return Avenger(self.name + " * " + other.name, self.power * other.power)

    def __truediv__(self, other):

        return Avenger(self.name + " / " + other.name, self.power / other.power)

    def __str__(self):

        return f"{self.name} with power {self.power}"
```

```python
thor=Avenger("Thor", 100)
hulk=Avenger("Hulk", 80)
```

```python
team = thor + hulk
```

```python
print(team)
```

```python
class AIModel:

    def __init__(self, name, accuracy):

        self.name = name

        self.accuracy = accuracy

    def __add__(self, other):

        return AIModel(self.name + " Ensemble " + other.name, (self.accuracy + other.accuracy) / 2)

    def __str__(self):

        return f"{self.name} with accuracy {self.accuracy}%"
```
