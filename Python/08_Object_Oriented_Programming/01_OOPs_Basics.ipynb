## Types of Constructor:
#### 1. Default Constructor
#### 2. Parameterised Constructor

-----------------------

#### 1. Default Contractor: It doesnt have the parameters inside it and it is only having the self keyword inside it. It assigns a default value to all of the objects.

```python
class Dog:
    def __init__(self):
        self.name="Labrador"
        self.color="Brown"

    def show_info(self):
        print(f"The dog's name is {self.name} and its color is {self.color}.")
```

```python
dog_1=Dog()
```

```python
dog_1.name
```

```python
dog_1.show_info()
```

------------

```python
dog_2=Dog()
```

```python
dog_2.name
```

-----------

#### 2. Parameterised Constructor : It consists of parameters inside it so that every object can have a specific attribute associated to it.
#### It allows us to pass arguments while creating an object.

```python
class Dog:
    
    def __init__(self,breed,color):
        self.breed=breed
        self.color=color
    def show_info(self):
        print(f"The dog's breed is {self.breed} and its color is {self.color}.")
```

```python
dog_1=Dog("Golden Retriever", "Golden")
```

```python
dog_1.show_info()
```

```python
dog_2=Dog("Husky", "Grey")
```

```python
dog_2.show_info()
```

#### 1. Instance Method:  Instance Methods operate Object atrribute.

```python
class Avengers:
    def __init__(self,name,power):
        self.name=name
        self.power=power

    def show_details(self):
        print(f"My name is {self.name} and My power is {self.power}.")
```

#### 2. Class Method

##### It is the method of the whole class or the whole blueprint which will remain the same for all of the objects then such kind of Method is being reffered as a Class Method

##### In class method, These methods work with the class level attributes.
##### They are going to use cls instead of self
##### They are called using the class name as well
##### For Specific that the current method is a class method we have to use the abbreviation as @class-methods.

```python
class Avengers:

    origin="Marvel Comics"
    
    def __init__(self,name,power):
        self.name=name
        self.power=power

    @classmethod
    def origin_info(cls):
        print(f"All Avengers are from {cls.origin}.")

    def self_info(self):
        print(f"My name is {self.name} and My power is {self.power}.")
```

```python
Iron_man=Avengers("Tony Stark","Genius Intellect and Technology")
```

```python
Iron_man.power
```

```python
Iron_man.origin_info()
```

```python
Iron_man.self_info()
```

```python
Thor=Avengers("Thor","Godly Strength and Control over Lightning")
```

```python
Thor.origin_info()
```

#### 3. Static Method

##### Static Methods do not use self or cls

##### They work like normal functions but belong to a class
##### They are going to be designed by a abbreviation @staticmethod
##### Static method can also be called by the classname also.

```python
class Avengerss:

    @staticmethod
    def mission():
        print("To protect the world from threats and save humanity.")
    
    origin="Marvel Comics"  

    def __init__(amit,name,power):
        amit.name=name
        amit.power=power

    def self_info(amit):
        print(f"My name is {amit.name} and My power is {amit.power}.")  

    @classmethod
    def origin_info(cls):
        print(f"All Avengers are from {cls.origin}.")

    def introduction(amit):
        print(f"My name is {amit.name}, I am an Avenger and My power is {amit.power}.")

    
```

```python
Iron_man=Avengerss("Tony Stark","Genius Intellect and Technology")
```

```python
Iron_man.introduction()
```

```python
Iron_man.origin_info()
```
