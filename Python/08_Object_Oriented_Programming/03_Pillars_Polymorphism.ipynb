# Polymorphism

### Polymorphism means MANY FORMS in GREEK.
### In OOPs, Polymorphism allows objects of different classes as if they are the object of the same class & it enables reusability and makes programs more flexible

## Types of Polymorphism
1. Method Overriding : ( Runtime Polymorphism )
2. Method Overloading : ( Compile Time Polymorphism )
3. Operator Overloading : ( Operator Overloading )

---

## Method Overriding
#### A child class provides a specific implementation of a method that has already been defined into the parent class .
This allowes Dynamic Selection at runtime.

```python
class Grandparent:  ## Parent class

    def speak(self):
        print("Hello My Name is Priyanshu.")

class Parent(Grandparent):  ## Child Class

    def speak (self):
        print("Hello My Name is bada Priyanshu and I am a parent.")

class Child(Grandparent):

    def speak(self):
        print("Hello My Name is chhota Priyanshu and I am a child.")
```

```python
def attribute_provider(objects):
    print(objects.speak())
```

```python
grandpa=Grandparent()
```

```python
attribute_provider(grandpa)
```

```python
parent_1=Parent()
```

```python
attribute_provider(parent_1)
```

```python
child=Child()
```

```python
attribute_provider(child)
```

```python
class Shape:
    def area(self):
        return "This method returns the area of the shape."
        
```

```python
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width=width
        self.height=height
            
    def area(self):
        return self.width*self.height

class Circle(Shape):
    def __init__(self, radius):
        self.radius=radius

    def area(self):
        return 3.14 * self.radius ** 2
```

```python
shape_1=Shape()
```

```python
shape_1.area()
```

```python
rectangle=Rectangle(6,4)
rectangle.area()
```

```python
circle=Circle(5)
```

```python
circle.area()
```
