# Abstraction

### Abstraction is the concept of hiding the implementation details and show the necessary features of the object. This helps in reducing programming complexity and efforts

### Abstraction allows us to work with an object with what an object does rather than how it does it

----------

### How can we achieve abstraction in python

#### 1. Abstract Class --> A class that cannot be instantiated or that means for an abstract class we can create the object of this class

#### 2. Abstract Method --> Abstract method just because it cant accessed by the object of the abstract class itself thats why we have to call this abstract method in our child class and the implementation of the abstract method is not provided inside the absract class but must be provided in the child class.
#### The abstract method is defined by @abstractmethod

------------

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def stop(self):
        pass

class Car(Vehicle):

    def start(self):
        print("Car is starting with a key.")

    def stop(self):
        print("Car is stopping with brakes.")

class ElectricScooter(Vehicle):

    def start(self):
        print("Electric Scooter is starting with a button.")

    def stop(self):
        print("Electric Scooter is stopping with a brake lever.")
```

```python
car = Car()
```

```python
car.start()
```

```python
car.stop()
```

```python
scooter = ElectricScooter()
```

```python
scooter.start()
```

```python
scooter.stop()
```
