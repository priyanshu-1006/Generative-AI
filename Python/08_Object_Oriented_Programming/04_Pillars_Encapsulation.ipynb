# Encapsulation

#### Encapsulation is a way of wrapping data (Variables and Methods) inside a capsule or a class as a Single Unit

#### It restricts direct access to some of the objects components which is a means of preventing accidental inference and misuse of the data

------------

#### Key Concepts of Encapsulation:

##### 1. Private Numbers.
##### 2. Protected Members
##### 3. Private Members
##### 4. Getter and setter methods

-----

#### Public Member ---> Accessible from Anywhere or attributes are accessible outside of the class and these members are accessible by the derived class.
Public Member attribute is directly declared



```python
class Ironman:
    def __init__(self,name,power,weapon):
        self.name = name
        self.power = power
        self.weapon = weapon

    def introduce(self):
        print(f"My name is {self.name}, I have {self.power} power and my weapon is {self.weapon}.")
```

```python
## dir()---> It takes the input as the object name and gives to us this idea that which member is public or not
```

```python
Iron=Ironman("Tony Stark","superhuman","suit")
```

```python
dir(Iron)
```

```python
Iron.name
```

```python
Iron.introduce()
```

### Private Members-->
##### A private member is not accessible outside the class and not even in the derived class and in order to have the access over these private members or to bring some modification on these members we use getter and setter method.

##### Private Members are declared by using double underscore __ before the name.

```python
class Avenger:
    def __init__(self,name,phone_number):
        self.name = name
        self.__phone_number = phone_number
```

```python
avenge=Avenger("Captain America","1234567890")
```

```python
avenge.name
```

```python
avenge.phone_number
```

```python
dir(avenge)
```

```python
## Gettter Function---> It is used to access the private member of the class
```

```python
## getter method is used to access the private member of the class. It is defined inside the class and it returns the value of the private member.
```

```python
class Captain:
    def __init__(self,name,address):
        self.name = name
        self.__address = address

## getter method 

    def accessing_address(self):
        print(f"Address of {self.name} is {self.__address}")
```

```python
captain=Captain("Captain America","New York")
```

```python
captain.name
```

```python
captain.address
```

```python
captain.__address
```

```python
captain.accessing_address()
```

```python
## Setter method---> It is used to modify the private member of the class. It is defined inside the class and it takes the value as input and modifies the value of def private member.
```

```python
class Animal:
    def __init__(self,name,sound):
        self.name = name
        self.__sound = sound

    def sound_info(self):
        print(f"{self.name} makes {self.__sound} sound.")

    def sound_modify(self,new_sound):
        self.__sound = new_sound
        print(f"{self.name} now makes {self.__sound} sound.")



```

```python
animal=Animal("Dog","Bark")
```

```python
animal.name
```

```python
animal.sound_info()
```

```python
animal.sound_modify("Woof")
```

## Protected Data

#### Protected Data is the data which is not accessible outside the class but the protected data is going to be accessible by the derived class or child class.

#### For having the protected member you can mention your protected member as _attribute or by using a single underscore

```python
class GrandParent:
    def __init__(self,name,locker_key):
        self.name = name
        self._locker_key = locker_key

        ## getter method
        def locker_info(self):
            print(f"{self.name} has locker key {self._locker_key}")
```

```python
class Parent(GrandParent):
    def __init__(self,name,locker_key,car_key):
        super().__init__(name,locker_key)
        self.__car_key = car_key

    def child_reveals(self):
        print(f"{self.name} has car key {self._locker_key}")
    ## getter method
    def car_info(self):
        print(f"{self.name} has car key {self.__car_key}")
```

```python
grandpa=GrandParent("Grandpa Joe","Locker123")
```

```python
grandpa.locker_key
```

```python
dir(grandpa)
```

```python
grandpa.locker_info()
```

```python
parent=Parent("Parent Mike","Locker456","Car789")
```

```python
parent.child_reveals()
```

```python
### Bank Account System
```

```python
class BankAccount:
    def __init__(self,account_holder,balance):
        self.account_holder = account_holder
        self.__balance = balance

    def get_balance(self):
        print(f"{self.account_holder} has a balance of {self.__balance} dollars.")

    def deposit(self,amount):
        if amount > 0:
            self.__balance += amount
            print(f"{amount} dollars deposited. New balance is {self.__balance} dollars.")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self,withdraw):
        if 0<=withdraw<=self.__balance:
            self.__balance -= withdraw
            print(f"{withdraw} dollars withdrawn. New balance is {self.__balance} dollars.")
        else:
            print("Invalid withdraw amount. Please check your balance.")
```

```python
account=BankAccount("Priyanshu",10000)
```

```python
account.get_balance()
```

```python
account.withdraw(2000)
```

```python
account.deposit(5000)
```
