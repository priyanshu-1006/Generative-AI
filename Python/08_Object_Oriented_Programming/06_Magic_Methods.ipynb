# Magic Methods

### Magic Methods are also called as Dunder Methods because they have double underscores __ before and after their name and they are special methods in python that allows us to customise how objects behave

### These Methods can help us define :
#### 1. Object Initialisation (__init__)
#### 2. String Representation (str,repr)
#### 3. Arithmetic Operations (add, sub etc.)
#### 4. Comparison Operations (eq, lt)
#### 5. Length, indexing, iteration (len,getitem, etc)

------------

## Why do we use Magic Methods

#### 1. Makes Object behave like built in types (eg, lists, integers)
#### 2. Improves Readability by allowing basic operations object1 +object2
#### 3. Simplifies code by replacing verbose function calls with natural syntax 



------

### print(object) --> str --> what is the returned when print(object) is used

#### Arithmetic operations can be done among the objects by using Magic Methods

```python
## __add__

class Avenger:

    def __init__(self,name,strength):
        self.name=name
        self.strength=strength

    def __add__(self,other):
        return Avenger(self.name + " & " + other.name, self.strength + other.strength)

    def __str__(self):
        return f"{self.name} has strength {self.strength}"

ironman = Avenger("Ironman", 85)
thor = Avenger("Thor", 95)

team=ironman + thor
print(team)
```
