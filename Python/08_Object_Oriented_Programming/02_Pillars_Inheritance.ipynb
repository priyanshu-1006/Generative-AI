## Inheritance

#### Inheritance is a Fundamental concept in Object Oriented Programming because of which child class is able to inherit the properties or attributes and functionalities of the parent class.

#### Just like a child is going to inherit the properties from parent and parent is going to inherit the properties from the frandparent and difinitely in this hierachy the grandchild can have some features different from the parents and grandparents as well.

-----

#### In Python typically we get to see 3 types of Inheritance
##### 1. Single Inheritance
##### 2. Multilevel Inheritance
##### 3. Multiple Inheritance

--------

#### 1. Single Inheritance
##### The features or the attributes from one parent class will be getting inherited into one child class.

##### Parent class (Properties<---->Attributes and Functionalities)
##### |
##### |
##### Child Class

```python
class Avenger:
    def __init__(self,name,power):
        self.name=name
        self.power=power
    
    def introduction(self):
        print(f"I am {self.name} and I have {self.power}.")

class IronMan(Avenger):

    def __init__(self,name,power,weapon):
        super().__init__(name,power)
        self.weapon=weapon

    def weapon_info(self):
        print(f"My weapon is {self.weapon}.")
```

```python
Iron=IronMan("Iron Man","Genius Intellect and Technology","Iron Suit")
```

```python
Iron.name
```

```python
Iron.weapon
```

```python
Iron.weapon_info()
```

```python
Iron.introduction()
```

### Multilevel Inheritance 

#### The properties or Methods and Attributes of the grandparent class can be inherited by the parent class and then the properties of the parent class will be inherited by the child class. In a proper hierarchy the Inheritance process is going to work for us.

### Grandparents
### |
### |
### |
### Parents
### |
### |
### |
### Child

```python
class Grandparent:

    def __init__(self,nationality,baldness,anger_issues):
        self.nationality=nationality
        self.baldness=baldness
        self.anger_issues=anger_issues

    def introduction(self):
        print(f"I am a {self.nationality} grandparent with {self.baldness} and {self.anger_issues}.")
```

```python
class Parent(Grandparent):

    def __init__(self,nationality,baldness,anger_issues,intelligence):
        super().__init__(nationality,baldness,anger_issues)
        self.intelligence=intelligence

    def intelligence_info(self):
        print(f"My intelligence level is {self.intelligence}.")
```

```python
class Child(Parent):
    def __init__(self,nationality,baldness,anger_issues,intelligence,charm):
        super().__init__(nationality,baldness,anger_issues,intelligence)
        self.charm=charm

    def charm_info(self):
        print(f"My charm level is {self.charm}.")
```

```python
parent_1=Parent("Indian","Bald","No","High")
```

```python
parent_1.introduction()
```

```python
parent_1.intelligence_info()
```

```python
child_1=Child("Indian","No","Yes","High","Very High")
```

```python
child_1.nationality
```

```python
child_1.charm_info()
```

```python
child_1.intelligence_info()
```

### Multiple Inheritance

#### In Multiple Inheritance, the properties of two parent class or multiple parent class will be getting inherited into the child class

### Parent_1 (Properties) + Parent_2 (Properties) + Parent_3 (Properties)....
### |
### |
### |c
### Child

```python
class Ironman:
    def __init__(self,name,power,weapon,nationality="American"):
        self.name=name
        self.power=power
        self.weapon=weapon
        self.nationality=nationality

    def introduction(self):
        print(f"I am {self.name} and I have {self.power}.")

class Captainamericas:
    def __init__(self,name,power,weapon,nationality="American"):
        self.name=name
        self.power=power
        self.weapon=weapon
        self.nationality=nationality

        def weapon_info(self):
            print(f"My weapon is {self.weapon}.")

class Hulk(Ironman,Captainamericas  ):
        def __init__(self,name,power,weapon,nationality="American",anger_issues="Yes"):
            super().__init__(name,power,weapon,nationality)
            self.anger_issues=anger_issues

        def anger_info(self):
            print(f"I have {self.anger_issues} anger issues.")
```
