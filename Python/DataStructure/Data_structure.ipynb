# Data Structures


#### Data structures in Python are being used for storing the data in a structured format so that we can access the data and along with that we can perform all of the operations with these data structures that we can perform in the case of databases as well like CRUD (Create, Read, Update, Delete) operations

## Types of Data Structures in Python are

1. List
2. Tuple
3. Set
4. Dictionary

The different types of data structures differs on this basis that in which specific format we want to store our data
