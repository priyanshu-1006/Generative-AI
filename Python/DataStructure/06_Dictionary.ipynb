## Keys()
It is a method in dictionary which returns all of the keys present inside our dictionary

```python
print(dictionary.keys()) # to get all the keys of the dictionary
```

```python
dictionary
```

## Values()
It will return all values inside our dictionary

```python
dictionary.values() # to get all the values of the dictionary
```

## items()
It will be used for returning the key value pairs together

```python
dictionary.items() # to get all the key value pairs of the dictionary in the form of a list of tuples
```

## get()
It will be used for returning the value for the key we are going to pass inside our get method

```python
dictionary.get("Name") # to get the value of a key using get() method
```
