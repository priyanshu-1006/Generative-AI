## List


List are one of the type of data structure which are used for sharing your data in a structured format where you can store our data of any data type and that as it is one of the data structure so we can have access to the data as well that we are going to store into the list and we can apply CRUD operations as well ie. create read update and delete operations as well


List are ordered and are a mutable Data Structure

# List Indexing

```python
## By using proper indexing value we can retrive the data from our lists
## indexing is from 0 to n-1 where n is the number of elements in the list

lists=[1,2,3,4,5,"Priyanshu Chaurasia",True]
print(lists[0]) # it will print 1
print(lists[1]) # it will print 2
print(lists[2]) # it will print 3   
```
