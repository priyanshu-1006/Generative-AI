# Streamlit

### Streamlit is an Open source Python Library used to build Beautiful, interactive web applications for data science and machine learning with minimal effort

### Why to use Streamlit

#### Easy to use
#### Fast Development
#### Interactive UI Elements ----> *Buttons, Sliders, Text Inputs, Charts, etc*
#### Great for Data Science ----> *Support Pandas, Matlplotib, Plotly and other libraries Deployment is easy*
