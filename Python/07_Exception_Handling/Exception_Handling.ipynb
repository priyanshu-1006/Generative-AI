#### Exception Handling is a way in which if we are getting the Errors in programming or if the user entered input causes an error, then instead of throwing an error we can raise an exception which can let the user know that the input is not correct or the given input can come up as an Error. As we dont want to stop the execution of the program, so we raise an error which will let the user know that the current input is Not Correct but the program execution wont stop.

-----------------------

#### Exception are events that disrupt the normal flow of a program. They occur when an error is encountered during program execution

-------------------------------

#### Different Types of Exception:

##### ZeroDivisionError ----> Dividing by Zero
##### FileNotFoundError ----> File Not Found
##### ValueError ----> Invalid Value
##### TypeError ----> Invalid Type

```python
# Exception Handling in Python

## Try and Except Block ---> It is used to handle exceptions that may occur during the execution of a program. The code that may raise an exception is placed inside the try block, and the code to handle the exception is placed inside the except block.
```

```python
### try :                                   [You are trying to run the operation that may raise an exception]
###         result=a/b

### except:                                 [You are going to represent that if the error has been thrown that should happen

###         print("The division is not valid")]
```

```python
try:
    a=b
except:
    print("Value b is not assigned")    
```

```python
try:
    a=b
except NameError as exception:
    print(exception)    
```

#### Try ---> Try block is going to have the operations where we can expect an error to occur
#### Except ---> Which is used to handle the error in a way such that the program execution shouldn't stop
#### else ---> It will be consisting of the code which we want to run if the try block operation is working without throwing any error
#### finally ---> Finally block gets executed once we are done with everything and even if we are getting an error or Not getting an error still the code inside the finally block  gets executed.

--------------------

### if try ----> code is throwing an error ---> Except ----> code will be runnning
### if try ----> code is not throwing an error ---> Else ---> Code will be running

```python
num = int(input("Enter a number: "))
try:
    result = 1 / num
except Exception as ex:
    print("An error occurred:", ex)

else:
    print("The division was successful, the result is:", result)
```
