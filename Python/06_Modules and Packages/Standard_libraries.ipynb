### JSON Library

```python
## When we are going to with NoSQL Databases data then the data inside the NoSQL databases is going to be in JSON format. So, we need to have knowledge of JSON library to work with that data. JSON library provides functions for working with JSON data, such as parsing JSON strings, converting Python objects to JSON format, and more. It is commonly used in applications such as web development, data exchange, and working with APIs.
```

```python
## When we get the JSON data, we get in a json string format. So by using json library we can convert our string data into a python dictionary format and serializing our data into json format.
```

```python
import json
```

```python
data = {"Name": "Priyanshu", "Age": 22, "City": "Varanasi"}
json_str=json.dumps(data)
print(json_str)
print(type(json_str))
```

```python
json_dict=json.loads(json_str)
print(json_dict)
print(type(json_dict))
```

### Regular Expression Library

```python
## Regular Expressions Library ---> It provides functions for working with regular expressions, which are a powerful tool for pattern matching and text manipulation. It is commonly used in applications such as data validation, text processing, and searching for specific patterns in strings.
```

```python
import re
```

```python
text = 'The rain in Spain stays mainly in the plain.'
```

```python
pattern=r'\d+'
match = re.search(pattern,text)
print(match.group())
```
