#### *Pandas Series is a column of data in a table like in Excel or Google Sheets with labels called as indexes attached to each value. It can hold any data type and is one-dimensional.*

```python
pd.Series((40, 50, 60), index=['Bread', 'Butter', 'Milk'], name='Groceries price')
```

```python
## Series is a 1D Array with labels (index) attached to each value. It can hold any data type int, float, string, etc. and is one-dimensional.
```

#### Data type --> Integers --> Data is a Series(column)---> Quantity of items in the stock
#### Data type --> Float --> Data in a series--> Price or Product Ratings
#### Data type --> Strings --> Column --> Product Name
#### Data type --> Boolean --> Column --> Stock or not
#### Data type --> Dates/time --> Column ---> Expiry Dates or sales dates
#### Mixed Type --> Milk,40,True

## 1. **Lists**

```python
## Series from List

country=['India', 'USA', 'UK', 'Germany', 'France']
pd.Series(country)
```

```python
batsman_runs = pd.Series([100, 150, 200, 250], index=['Virat', 'Rohit', 'Dhoni', 'Kohli'], name='Runs scored')
pd.Series(batsman_runs, dtype='int32')
```

```python
subject_names = pd.Series(['Maths', 'Science', 'English', 'History'], name='Subjects')
subject_marks= pd.Series([90, 85, 95, 80], index=['Maths', 'Science', 'English', 'History'], name='Marks')
pd.Series(data=subject_marks, index=subject_names, name= 'Subject Marks Data', dtype='int32')
```

```python
pd.Series(subject_marks)
```

----

## 2. **Numpy Array**

```python
### Creating a Series using a Numpy array
values=np.arange(10,21)
pd.Series(values)
```

------

## 3. **Dictionary**

```python
## Dictionary to Series

## Dictionary is another data structure in Python which is used to store data in key-value pairs. We can create a Series from a dictionary where the keys become the index and the values become the data of the Series.
## Just becuase it stores the data in key value pair so we dont have to explicitly mention the index and data while creating a Series from a dictionary. It will automatically take the keys as index and values as data.
```

```python
data_items = {'Maths': 90, 'Science': 85, 'English': 95, 'History': 80}
pd.Series(data_items, name='Subject Marks')
```

## Mathematical Functions in Pandas

### 1. **sum()**
#### *Sum of all values present in the series*

### 2. **product()**
#### *Product of all values in the series.*

### 3. **mean()**
#### *Mean provides us with average value of the data column*

### 4. **median()**
#### *Median also provides the idea about the average of the data, for coming up with median, we need to SORT the values in Ascending manner and then the middle value after sorting the data* 

### 5. **mode()**
#### *Mode provides us with the idea that which value is getting repeated maximum number of times.*

### 6. **std()** -
#### *If I need to get the idea that how the values inside the series are distributed from the average or mean value then we can come up with the standard deviation*

### 7. **variance()**
#### *Variance also provides the insight about the distribution of the data from the value and variance is the square of the standard deviation*

### 8. **min()**
#### *minimum value in the series*

### 9. **max()**
#### *maximum value in the series*

### 10. **describe()**
#### describe fn provides the idea about the summary of the important mathematical calculations for the data

### 11. **count**

```python
## Indexing in Series

#### 1. Positive Indexing (Just like List Indexing) --> Start from 0 to n-1

#### 2. No Negative Indexing

#### 3. Slicing --> In Series we can get the certain values over a given specific range and we can do it using [initial_val, final_val] --> It will return the values from initial_val to final_val-1

#### 4. Fancy Indexing --> It is a technique where we can get the certain values from a Series by passing a list of index values. It will return the values corresponding to the index values passed in the list.

#### 5. Indexing with Labels --> A series which is consisting of Label as the index or custom index, then for this data we can retrieve the info using that custom index or label instead of the default index.
```
