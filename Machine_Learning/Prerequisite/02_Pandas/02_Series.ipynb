#### *Pandas Series is a column of data in a table like in Excel or Google Sheets with labels called as indexes attached to each value. It can hold any data type and is one-dimensional.*

```python
pd.Series((40, 50, 60), index=['Bread', 'Butter', 'Milk'], name='Groceries price')
```

## 1. **Lists**

```python
## Series from List

country=['India', 'USA', 'UK', 'Germany', 'France']
pd.Series(country)
```

```python
batsman_runs = pd.Series([100, 150, 200, 250], index=['Virat', 'Rohit', 'Dhoni', 'Kohli'], name='Runs scored')
pd.Series(batsman_runs, dtype='int32')
```

```python
subject_names = pd.Series(['Maths', 'Science', 'English', 'History'], name='Subjects')
subject_marks= pd.Series([90, 85, 95, 80], index=['Maths', 'Science', 'English', 'History'], name='Marks')
pd.Series(data=subject_marks, index=subject_names, name= 'Subject Marks Data', dtype='int32')
```

```python
pd.Series(subject_marks)
```

```python
### Creating a Series using a Numpy array
values=np.arange(10,21)
pd.Series(values)
```

## 3. **Dictionary**

```python
## Dictionary to Series

## Dictionary is another data structure in Python which is used to store data in key-value pairs. We can create a Series from a dictionary where the keys become the index and the values become the data of the Series.
## Just becuase it stores the data in key value pair so we dont have to explicitly mention the index and data while creating a Series from a dictionary. It will automatically take the keys as index and values as data.
```

```python
data_items = {'Maths': 90, 'Science': 85, 'English': 95, 'History': 80}
pd.Series(data_items, name='Subject Marks')
```

```python
## Loop in Series
#### We can loop through a Series using a for loop. We can access the values of the Series using the index or we can access the values directly. We can also use the iteritems() method to loop through the Series and get both the index and the value at the same time.
for value in Netflix['Genre']:
    print(value)
```

```python
## Arithmetic Operators in Series
## We can perform arithmetic operations on Series just like we do with numbers. The arithmetic operations are performed element-wise, which means that the operation is performed on each element of the Series. The result of the operation is a new Series with the same index as the original Series.
Series1 = pd.Series([1, 2, 3, 4, 5])
Series2 = pd.Series([10, 20, 30, 40, 50])
Series1 + Series2
```

```python
Series1 - Series2
```

```python
addedData = Series1 + 1000
```

```python
addedData
```

```python
50 - addedData
```

```python
## Relational Operators in Series

## We can perform relational operations on Series just like we do with numbers. The relational operations are performed element-wise, which means that the operation is performed on each element of the Series. The result of the operation is a new Series with the same index as the original Series, but with boolean values (True or False) indicating the result of the relational operation for each element.
```

```python
400 > addedData
```

```python
400 < addedData
```

## Boolean Indexing

#### In order to perform the Important analysis on the column values, where in a specific index value If it is meeting a condition that we get True over their else we get False when the condition is not meeting for that specific index.

###### Boolean Index - True -> Condition is true
###### Boolean Index - False -> Condition not true.

```python
### Which Youtubers are having average video length greater than 20 minutes?

average_video_length = pd.Series([15, 25, 10, 30, 20], index=['Youtuber1', 'Youtuber2', 'Youtuber3', 'Youtuber4', 'Youtuber5'])
average_video_length > 20
```
