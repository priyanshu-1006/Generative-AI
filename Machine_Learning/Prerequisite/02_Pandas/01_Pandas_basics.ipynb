# **Pandas**

-----

### *Pandas is a Powerful Library used for Data Manipulation and Analysis. It helps us to work with data in Tabular Format, just like how we deal with data and spreadsheet in excel or database.*

--------

### Pandas provides functionalities and attributes to clean organise and analyse large amounts of data

---

#### 1. *Easy To Use* - Works like excel inside Pythom
#### 2. *Fast* - Handles large data faster than manual methods.
#### 3. *Clean Data* - helps clean, filter, and transform messy data
#### 4. *Integrates well* - Matplotib, SeaBorn
#### 5. *Ideal for ML and AI* - It is ideal for data science.
