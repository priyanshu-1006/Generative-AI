## **Data Exploration** -
#### Attributes & Functions for the dataframe

### Attributes
#### 1. **Shape** - It provides the info about the (row, column) information about the dataset
##### Syntax - df.shape

#### 2. **dtypes** - It provides the idea about the datatype of all columns , so all columns with their name and their associated data type will be shown by using dtypes.
##### Syntax - df.dtypes

#### 3. **index** - For a dataframe with labelled index we can get the ideaa about the label just by using index attribute.
##### Synatx - df.index

#### 4. **Columns** - It provides the idea about all of the column names of the given dataframe

#### 5. **values** - It provides the idea about the values present row by row inside the dataframe. It always return a numpy array representing the values of column row by row.

#### 6. **ndim** - It provides the idea about the dimension of the dataframe.

#### 7. **size** - It provides us idea about Total number of elements present inside the dataframe, that means all rows * all columns (Column 1 - 50 rows , Column 2 - 50 rows , Column 3 - 70 rows,  --> size - 50 + 50 + 70)

# **Functions**

#### 1. **head()** - To get to have the idea about the data that your dataset is storing as the top 5 rows, that means that starting from the top we can get the top 5 rows and along with that if I want that instead of top 5 rows i want to see a fixed number of rows from the rows, that number i can pass inside the head() as a parameter

#### 2. **tail()** - Provide the insight about the dataframe from the bottom so tail() can show us bottom 5 rows and aagin based on the value you will be passing as the parameter, those many values you can get using the tail fn()

#### 3. **sample()** - It provides the insight about the dataframe as the random values present inside the dataset and again based on how many values I want i can put the parameter as per that. 

#### 4. **info()** - Info function provides the idea about 
##### 1. *the datatypes of all columns*
##### 2. *Number of Missing values of all columns*
##### 3. *DataFrame Storage in Memory*

#### 5. **describe()** - It provides an idea as the mathematical summary of the data -
##### minimum, maximum, count, mean, std, 25%, 50%, 75%
