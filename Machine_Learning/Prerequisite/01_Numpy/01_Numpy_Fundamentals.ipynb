# **NUMPY**

##### **Numpy is a Fundamental Package for scientific computing in Python.** *It is a Python Library that provides a Multi Dimensional Array Object and various derived objects such as Masked Arrays or Matrices and an assortment of routines for fast operations on Arrays including Mathematical, Logical, Shape Manipulation, Sorting, Selecting, I/O, Discrete Fourier Transform, Basic Linear Algebra, Statistical Operations, Random Simulations,etc.

##### At the core of the Numpy Package is the n dimensional Array object.



## Why Numpy ?

### 2008-2009, Data Science ----> Matlab, R language

## Data Types in Python ---> **List, Dictionary --> Slow**

### *Syntax --> Python --> Easy*
### We needed speed and less memory to work with the huge amount of data.

### *Numpy ---> C Language ---> Wrapped ---> Python Language*
### *Numpy ---> Syntax ---> Easy for Beginners --> Performing Scientific Calculations----> Creating Multi Dimensional Arrays---> Speed is faster than applying Operations ---> Less Memory*

### Libraries like **Scikit Learn, Tensorflow and PyTorch** require input data in Numpy Arrays or Tensors.

## **Numpy Arrays** Vs **Python Sequences** 

#### Numpy Arrays have a fixed size at creation, unlike Python Lists which grow dynamistically. Chaning the size 0f an n dimentional array will create a new array and delete the original.

#### The elements in a Numpy Array are required to be of the same data type and thus will be of the same size in memory.

#### Numpy Arrays facilitate advanced Mathematical and other types of operations on large number of data. Typically, such operations are executed more efficiently with less code than is possile using python's built in sequences.
