### Types of Plots

##### 1. Line Plot ---> Time Series analysis --> How values are varies as per time
##### 2. Scatter Plot ---> Scatter Plot is being used for Bivariate Analysis --> Two Numerical Columns --> we check that based on the changes in one Numerical Column Value --> How it affects the other numerical Column value.
##### for example :- x--> Age ---> 10,15,20,25,30,..75
##### y ---> Cholestrol Level --> 50,60,75,80
##### Scatter Plot --> It will figure out that based on the increment in Age Value --> If Cholestrol level is Increasing or Decreasing or is Steady.

```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
df=pd.read_csv('Student_Depression_Dataset.csv')
```

```python
df.head(5)
```

```python
df['Age'].value_counts()
```

```python
### Age vs Average CGPA

plt.scatter(df['Age'], df['CGPA'])
plt.xlabel('Age')
plt.ylabel('Average CGPA')
plt.title('Age vs Average CGPA')
```

```python
data=df.groupby('Age')['CGPA'].mean().head(10)
```

```python
x=data.index
```

```python
y=data.values
```

```python
### Function to create a scatter plot ---> plt.scatter(x, y)
plt.title('Age vs Average CGPA')
plt.xlabel('Age')
plt.ylabel('Average CGPA')
plt.scatter(x, y, color='red', marker='*', s=100, alpha=0.5)
plt.grid()
plt.show()
```

```python
sales=pd.read_csv('Chocolate_Sales.csv')
```

```python
sales.sample(5)
```

```python
x=sales['Boxes Shipped']
y=sales['Amount']
```

```python
plt.scatter(x, y, color='red', marker='*', s=100, alpha=0.5)
```

```python
plt.figure(figsize=(10, 4))
plt.title('Total Amount of Variation as per Total Boxes Shipped')
plt.xlabel('Boxes Shipped')
plt.ylabel('Amount')
plt.scatter(x, y, color='red', marker='*', s=100, alpha=0.5)
plt.show()
```
