## **Customisation of line**
-----------
### Parameters
#### 1. **color** --> color of the line (red, blue, green, yellow, cyan, magenta, black, white, etc.)
#### 2. **marker** --> shape of the marker (o, s, ^, v, <, >, *, +, x, D, h, H)
#### 3. **linestyle** --> style of the line (solid, dashed, dashdot, dotted)
#### 4. **linewidth** --> width of the line (default is 1)
#### 5. **markersize** --> size of the marker (default is 6)
#### 6. **alpha** --> transparency of the line (default is 1)
#### 7. **label** --> label of the line (for legend) It may help when there are multiple lines for single visualisation.
#### 8. **grid** --> grid of the graph (default is False)

```python
## Line Plot -->  AGE vs CGPA
```

```python
data=df.groupby('Age')['CGPA'].mean().head(10)
```

```python
x=data.index
y=data.values
```

```python
plt.figure(figsize=(10,5))
plt.title('Average CGPA as per Age', fontsize=20, color='blue')
plt.ylabel('Average CGPA', fontsize=15, color='red')
plt.xlabel('Age', fontsize=15, color='green')
plt.plot(x,y, color='red', marker='o', linestyle='dashed', linewidth=2, markersize=10, label='Line Graph', markerfacecolor='blue', markeredgecolor='black', markeredgewidth=2, alpha=0.7,)
plt.grid()
plt.show()
```
