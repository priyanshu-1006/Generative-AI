##### **Scatter Plot** --> *Scatter plot is used for representing the relationship between two variables.*

```python
### Dietary Habits vs Depression
```

```python
df['Dietary Habits'].value_counts()
```

```python
df=df[df['Dietary Habits'] != 'Others']
```

```python
df['Dietary Habits'].value_counts()
```

```python
data=df[(df['Dietary Habits'] == 'Unhealthy') | (df['Dietary Habits'] == 'Moderate') | (df['Dietary Habits'] == 'Healthy')]
```

```python
data['Dietary Habits'].value_counts().plot(kind='bar',color=['red','yellow','green'])
```

```python
### remove_unused_categories() --> remove unused categories from categorical data
```

```python
df['Dietary Habits'].value_counts()
```

```python
data['Dietary Habits'].value_counts()
```

```python
d=data.groupby('Dietary Habits')['Depression'].sum()
```

```python
d
```

```python
x=d.index
y=d.values
```

```python
plt.bar(x,y,color=['red','yellow','green']) 
```

```python
### Gender ---> Distribution vs Academic Pressure
```

```python
data['Gender'].value_counts()
```

```python
Male=data[data['Gender']=='Male'].head(12348)
```

```python
Female=data[data['Gender'] =='Female'].head(12348)
```

```python
## concat() ---> concat() function joins two dataset or multiple dataset by using concat()
## pd.concat([df_1,df_2,df_3,.....]) 
```

```python
df=pd.concat([Male,Female])
```

```python
df['Gender'].value_counts()
```

```python
data=df.groupby('Gender')['Academic Pressure'].sum()
```

```python
data
```

```python
x=data.index
y=data.values
```

```python
plt.figure(figsize=(10,6))
plt.xlabel('Gender')
plt.ylabel('Academic Pressure')
plt.title('Gender vs Academic Pressure')
plt.bar(x,y,color=['red','yellow'])
```
