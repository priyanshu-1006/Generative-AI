```python
import matplotlib.pyplot as plt

# 1. Filter the data for only the three products you want
products = ['Mint Chip Choco', '85% Dark Bars', 'Peanut Butter Cubes']
df_filtered = sales[sales['Product'].isin(products)]

# 2. Pivot the data: Dates become rows, Products become columns
# fill_value=0 ensures missing months get a 0 instead of breaking the chart lengths
pivot_sales = df_filtered.pivot_table(index='Date', 
                                      columns='Product', 
                                      values='Amount', 
                                      aggfunc='sum', 
                                      fill_value=0)

# 3. Plot directly from the pandas dataframe
# pandas automatically handles side-by-side grouping and the legend!
pivot_sales.plot(kind='bar', width=0.7, figsize=(10, 6), color=['red', 'blue', 'green'])

# 4. Add labels and formatting
plt.title('Monthly Sales')
plt.xlabel('Months')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Stacked Bar Chart
### Instead of keeping the bars in side by side direction, we can keep the bars in the stacked manner i.e, One on top of other and in order to figure out which category is having maximum sale, we can have a look over the length of the Bar.

## Stacked Bar Chart --> bar() --> parameter --> bottom --> This bottom parameeter specific that what bar I want to keep below my current bar.

Red Bar
Green Bar --> Bottom --> Red Bar
Blue Bar --> Bottom --> Green Bar + Red Bar

## As we want to differentiate on the basis of Bar length Value so thats why we have to keep y_value of previous bar at the bottom.

```python
import matplotlib.pyplot as plt

# 1. Filter the data
products = ['Mint Chip Choco', '85% Dark Bars', 'Peanut Butter Cubes']
df_filtered = sales[sales['Product'].isin(products)]

# 2. Pivot the data (same as before)
pivot_sales = df_filtered.pivot_table(index='Date', 
                                      columns='Product', 
                                      values='Amount', 
                                      aggfunc='sum', 
                                      fill_value=0)

# 3. Plot a STACKED bar chart by adding stacked=True
pivot_sales.plot(kind='bar', stacked=True, width=0.7, figsize=(10, 6), color=['red', 'blue', 'green'])

# 4. Add labels and formatting
plt.title('Monthly Sales by Product')
plt.xlabel('Months')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)

# Move the legend outside the chart so it doesn't cover your bars
plt.legend(title='Product', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
